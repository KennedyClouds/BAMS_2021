```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import matplotlib.patheffects as path_effects

BASE_DIR = 'pictures/'
files = ['OSCRE_2020.jpg','OSCRE_GEN1.jpg','OSCRE_2021.jpg','OSCRE_GEN2.jpg',]
labels = ['(a)','(b)','(c)','(d)']
clabels = ['black','black','black','black']
#titles = ['1537','1737','1937','2137','2217','2312']
fig = plt.figure(figsize=(8,6),dpi=300)
for i in range(1, 5):
    fig.add_subplot(2, 2, i)
    img=mpimg.imread(BASE_DIR+files[i-1])
    imgplot = plt.imshow(img)
    plt.axis('off')
    text = plt.text(10, 200, labels[i-1], fontsize=12,weight='bold',color=clabels[i-1])
    #plt.title(titles[i-1]+' UTC - 24 Feb. 2019',fontweight="bold",fontsize=12)
    if i == 1:
        text.set_path_effects([path_effects.Stroke(linewidth=2, foreground='white'),
                       path_effects.Normal()])     
plt.subplots_adjust(hspace = 0.04, wspace = 0.01)
plt.savefig('OSCRE_HW.png', dpi=300, bbox_inches='tight')
#plt.savefig('outreach', format='pdf', bbox_inches='tight')
plt.show()
```
