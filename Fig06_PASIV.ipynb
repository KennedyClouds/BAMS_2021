```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import matplotlib.patheffects as path_effects

BASE_DIR = 'pictures/'
files = ['PASIV_inflate.jpg','PASIV_tree.jpg','PASIV_recovery.jpg','PASIV_data.png',]
labels = ['(a)','(b)','(c)','(d)']
clabels = ['white','black','white','black']
#titles = ['1537','1737','1937','2137','2217','2312']
fig = plt.figure(figsize=(8,6),dpi=300,facecolor='white')
for i in range(1, 5):
    fig.add_subplot(2, 2, i)
    img=mpimg.imread(BASE_DIR+files[i-1])
    imgplot = plt.imshow(img)
    plt.axis('off')

    if i < 4:
        text = plt.text(10, 260, labels[i-1], fontsize=12,weight='bold',color=clabels[i-1])
    if i == 4:
        text = plt.text(10, -250, labels[i-1], fontsize=12,weight='bold',color=clabels[i-1])
    #plt.title(titles[i-1]+' UTC - 24 Feb. 2019',fontweight="bold",fontsize=12)
    if i == 1:
        text.set_path_effects([path_effects.Stroke(linewidth=2, foreground='black'),
                       path_effects.Normal()])    
plt.subplots_adjust(hspace = 0.04, wspace = 0.01)
plt.savefig('Fig06_PASIV.png', dpi=300, bbox_inches='tight')
plt.savefig('Fig06_PASIV.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
