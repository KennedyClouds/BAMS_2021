This code produces Fig. 2 of Kennedy et al. (2021) in BAMS. It takes images from OSCRE and adds annotation. 

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
#import numpy as np
#each panel is 1936 x 1216
file = 'pictures/OSCRE_blsn.png'
labels = ['(a)','(b)','(c)','(d)']

fig = plt.figure(figsize=(6,8),dpi=300)
img=mpimg.imread(file)
imgplot = plt.imshow(img)
plt.axis('off')
plt.text(10, 70, '(a)', fontsize=12,weight='bold',color='white')
plt.text(10, 70+1216, '(b)', fontsize=12,weight='bold',color='white')
    #plt.title(titles[i-1]+' UTC - 24 Feb. 2019',fontweight="bold",fontsize=12)
plt.text(1380, 1180, '1 cm', fontsize=10,weight='bold',color='white')
plt.text(1380, 1180+1216, '1 cm', fontsize=10,weight='bold',color='white')

#plt.savefig('Fig02_OSCRE_BLSN.png', dpi=300, bbox_inches='tight')
#plt.savefig('Fig02_OSCRE_BLSN.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
